Gaussian noise is also called electronic noise, beucase it arises in amplifiers. 
effect of standard deviation(measure of amount of dispersion) also called sigma is directly prop to dots.
more the value of sigma more the dots on the image. 

```python
import cv2
from skimage import util
from skimage.color import rgb2gray
from matplotlib import pyplot as plt
from PIL import Image
#Open image using Image module (pillow)
import numpy as np
from skimage import color
from skimage import io
from skimage.util import random_noise
```

```python
#img1 = io.imread("zahoor.jpeg")
img1 = skimage.io.imread(fname="zahoor.jpeg", as_gray=True)
#img1 = color.rgb2gray(img2)
fig, ax = plt.subplots()
plt.imshow(img1, cmap="gray")
```

subplot:
It can be really useful to split your graphic window into several parts in order to display several charts at the same time. With matplotlib, this can be done using subplot() function. It is possible to display your graphics in several rows or several columns, or both. 

```python
img = np.array(img1)
#the default mean is 0 and variance is 0.1
noisy1 = util.random_noise(img, mode='gaussian', mean=0, var=0.01)
noisy2 = util.random_noise(img, mode='gaussian', mean=0.1, var=0.01)
noisy3 = util.random_noise(img, mode='gaussian', mean=0, var=0.2)
plt.subplot(221)

plt.title('original')
plt.xticks([]) # remove ticks
plt.yticks([])
plt.imshow(img1,cmap="gray")
#cmap=colormap
plt.subplot(222)
plt.title('mean=5, var=0.01')
plt.xticks([]) # remove ticks
plt.yticks([])
plt.imshow(noisy1 ,cmap="gray")
plt.subplot(223)
plt.title('mean=0.1, var=0.025')
plt.xticks([]) # remove ticks
plt.yticks([])
plt.imshow(noisy2 ,cmap="gray")
plt.subplot(224)
plt.title('mean=0, var=0.2')
plt.xticks([]) # remove ticks
plt.yticks([])
plt.imshow(noisy3 ,cmap="gray")
plt.show()
```
