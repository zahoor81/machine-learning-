```python
import cv2
from matplotlib import pyplot as plt
from skimage.util import random_noise
```

```python
img1 =cv2.imread("zahoor.jpeg",0)
#img1 = color.rgb2gray(img2)
fig, ax = plt.subplots()
plt.imshow(img1, cmap="gray")
```

```python
gauss=random_noise(img1,mode='gaussian',seed=None,clip=True)
salt=random_noise(img1,mode='salt',seed=None,clip=True)
pepper=random_noise(img1,mode='pepper', seed=None,clip=True)
```

```python
plt.subplot(221),plt.imshow(img1,cmap="gray"),plt.title('original')
plt.xticks([]) # remove ticks
plt.yticks([])
plt.subplot(222),plt.imshow(gauss,cmap="gray"),plt.title('gaussian')
plt.xticks([]) # remove ticks
plt.yticks([])
plt.subplot(223),plt.imshow(salt,cmap="gray"),plt.title('salt')
plt.subplot(224),plt.imshow(pepper,cmap="gray"),plt.title('pepper')
plt.show()
```
