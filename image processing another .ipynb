resizing the image 

```python
import skimage.transform
#actually compression is being done here 
image = skimage.io.imread("zahoor.jpeg")
# resize the image
new_shape = (image.shape[0] // 5, image.shape[1] // 5, image.shape[2])
#image.shape represents the image is 2d
small = skimage.transform.resize(image=image, output_shape=new_shape)
small = skimage.img_as_ubyte(small)
# write out image
skimage.io.imsave(fname="zahoor_res.jpeg", arr=small)
# display images
fig, ax = plt.subplots()
plt.imshow(image)
fig, ax = plt.subplots()
plt.imshow(small)
```

Thresholding: Looking through the array for pixel colour values that are less than some threshold value. 

```python
image=skimage.io.imread("zahoor.jpeg")
fig, ax=plt.subplots()
plt.imshow(image)
```

```python
# keep only high-intensity pixels
image[image < 128] = 0

# display modified image
fig, ax = plt.subplots()
plt.imshow(image)
```

converting colored image to grayscale 

```python
import skimage.color
```

```python
im=skimage.io.imread("zahoor.jpeg")
fig, ax=plt.subplots()
plt.imshow(im)
```

```python
# convert to grayscale and display
gray_image = skimage.color.rgb2gray(image)
fig, ax = plt.subplots()
plt.imshow(gray_image, cmap="gray")
```

```python
ima=skimage.io.imread("zahoor.jpeg",as_gray=True)
fig, ax=plt.subplots()
plt.imshow(ima,cmap="gray")
```
